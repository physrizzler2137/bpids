# 1. Basic Python Variables and Arithmetic

```python
speed_kmh = 50
time_s = 10.0

distance_m = speed_kmh / 3.6 * time_s
distance_m
```

# 2. Lists and Simple Operations
Lists are Python’s built‑in container type.

```python
speeds_kmh = [0, 20, 40, 60, 80]
speeds_ms = [v / 3.6 for v in speeds_kmh]
speeds_ms
```

# 3. NumPy Arrays
NumPy arrays allow vectorized operations (fast, efficient).

```python
import numpy as np

speeds_kmh = np.array([0, 20, 40, 60, 80])
speeds_ms = speeds_kmh / 3.6
speeds_ms
```

# 4. First Plot with Matplotlib
This is your first engineering visualization.

> **Reuse from earlier:** This plotting pattern will be reused in the mini‑project.

```python
import matplotlib.pyplot as plt

time_s = np.arange(0, 10, 0.5)
speed_kmh = 10 * time_s  # synthetic example

plt.figure(figsize=(8, 4))
plt.plot(time_s, speed_kmh)
plt.xlabel("Time [s]")
plt.ylabel("Speed [km/h]")
plt.title("Example Speed Profile")
plt.grid(True)
plt.show()
```
