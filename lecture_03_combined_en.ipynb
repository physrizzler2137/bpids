# Synthetic Dataset Generator

```python
import numpy as np
import pandas as pd

N = 2000
time_s = np.arange(N)

speed_kmh = np.abs(50 + 20*np.sin(time_s/50) + np.random.normal(0, 5, N))
rpm = np.abs(2000 + 500*np.sin(time_s/30) + np.random.normal(0, 100, N))

df = pd.DataFrame({
    "time_s": time_s,
    "speed_kmh": speed_kmh,
    "rpm": rpm
})

df.head()
```

# 1. Inspecting Data

```python
df.info()
```

# 2. Boolean Filtering

```python
df_fast = df[df["speed_kmh"] > 80]
df_fast.head()
```

# 3. GroupBy Aggregation

```python
df["minute"] = (df["time_s"] // 60).astype(int)

summary = df.groupby("minute").agg({
    "speed_kmh": "mean",
    "rpm": "mean"
})

summary.head()
```

# 4. Feature Extraction

```python
speed_ms = df["speed_kmh"].values / 3.6
time_s = df["time_s"].values

dt = np.mean(np.diff(time_s))
accel = np.diff(speed_ms) / dt

accel[:10]
```

# 5. Basic Classification Example

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

y = (df["speed_kmh"] > 60).astype(int).values
X = df[["speed_kmh", "rpm"]].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf = RandomForestClassifier()
clf.fit(X_train, y_train)

pred = clf.predict(X_test)
accuracy_score(y_test, pred)
```
