# 2. Listy i proste operacje
Listy to podstawowy typ kontenera w Pythonie.

```python
speeds_kmh = [0, 20, 40, 60, 80]
speeds_ms = [v / 3.6 for v in speeds_kmh]
speeds_ms
```

# 3. Tablice NumPy
Tablice NumPy umożliwiają szybkie operacje wektorowe.

```python
import numpy as np

speeds_kmh = np.array([0, 20, 40, 60, 80])
speeds_ms = speeds_kmh / 3.6
speeds_ms
```

# 4. Pierwszy wykres w Matplotlib
To Twój pierwszy wykres inżynierski.

> **Wykorzystasz ponownie:** Ten schemat rysowania wykorzystasz w mini‑projekcie.

```python
import matplotlib.pyplot as plt

time_s = np.arange(0, 10, 0.5)
speed_kmh = 10 * time_s  # przykład syntetyczny

plt.figure(figsize=(8, 4))
plt.plot(time_s, speed_kmh)
plt.xlabel("Czas [s]")
plt.ylabel("Prędkość [km/h]")
plt.title("Przykładowy profil prędkości")
plt.grid(True)
plt.show()
```
