# 1. Tablice NumPy i operacje wektorowe

```python
import numpy as np

speed_ms = np.array([0, 1, 2, 3, 4])
speed_ms
```

# 2. Obliczanie przyspieszenia za pomocą `np.diff`
Przyspieszenie to pochodna prędkości.

> **Wykorzystasz ponownie:** Ten schemat pojawi się w mini‑projekcie.

```python
dt = 1.0  # załóżmy próbkę co 1 sekundę
accel = np.diff(speed_ms) / dt
accel
```

# 3. Obliczanie jerku (pochodna przyspieszenia)

```python
jerk = np.diff(accel) / dt
jerk
```

# 4. Wygładzanie sygnału — średnia krocząca
> **Wykorzystasz ponownie:** Funkcja pojawi się w mini‑projekcie.

```python
def moving_average(x, window):
    return np.convolve(x, np.ones(window)/window, mode="valid")

moving_average(speed_ms, window=3)
```

# 5. Dashboard 2×2
Ten układ wykorzystasz w mini‑projekcie.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].plot(speed_ms)
axes[0, 0].set_title("Prędkość [m/s]")

axes[0, 1].plot(accel)
axes[0, 1].set_title("Przyspieszenie [m/s²]")

axes[1, 0].plot(jerk)
axes[1, 0].set_title("Jerk [m/s³]")

axes[1, 1].plot(moving_average(speed_ms, 3))
axes[1, 1].set_title("Prędkość wygładzona")

for ax in axes.ravel():
    ax.grid(True)

plt.tight_layout()
plt.show()
```
