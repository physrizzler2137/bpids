# 1. NumPy Arrays and Vectorized Operations

```python
import numpy as np

speed_ms = np.array([0, 1, 2, 3, 4])
speed_ms
```

# 2. Computing Acceleration with `np.diff`
Acceleration is the derivative of speed.

> **Reuse from earlier:** This pattern will be used in the mini‑project.

```python
dt = 1.0  # assume 1 second sampling
accel = np.diff(speed_ms) / dt
accel
```

# 3. Computing Jerk (Derivative of Acceleration)

```python
jerk = np.diff(accel) / dt
jerk
```

# 4. Moving Average Smoothing
A simple low‑pass filter.

> **Reuse from earlier:** This function will be reused in the mini‑project.

```python
def moving_average(x, window):
    return np.convolve(x, np.ones(window)/window, mode="valid")

moving_average(speed_ms, window=3)
```

# 5. Multi‑Plot Dashboard (2×2)
This layout will be reused in the mini‑project.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].plot(speed_ms)
axes[0, 0].set_title("Speed [m/s]")

axes[0, 1].plot(accel)
axes[0, 1].set_title("Acceleration [m/s²]")

axes[1, 0].plot(jerk)
axes[1, 0].set_title("Jerk [m/s³]")

axes[1, 1].plot(moving_average(speed_ms, 3))
axes[1, 1].set_title("Smoothed Speed")

for ax in axes.ravel():
    ax.grid(True)

plt.tight_layout()
plt.show()
```
